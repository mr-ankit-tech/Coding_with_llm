# **Loading the pdf into the FAISS vector db after embedding the data**

```python
# We start by fetching a document that loads the book - 'Atomic habits'
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings



pdfloader = PyPDFLoader(file_path=r"/content/Atomic Habits James Clear.pdf")
pages = pdfloader.load_and_split()

# Load the documents using LangChain's built in extractors, formatters,loaders,embeddings and LLMs
mydocuments = pdfloader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
text = text_splitter.split_documents(mydocuments)

# Specify a common and widely used Sentence Transformer model
embeddings = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)

#Using Langchain API functions to interact with FAISS
db = FAISS.from_documents(text, embeddings)
```

# **Creating a Retriver tool using Langchain where it will club the Vector DB, LLM**

```python
retriever = db.as_retriever()
from langchain.agents.agent_toolkits import create_retriever_tool
tool = create_retriever_tool(
    retriever,
    "My_Fav_Book_Atomic_Habits",
    "Searches and returns documents regarding the book atomic habits"
)
tools = [tool]

#Wrapping LLM with conversational agent
from langchain.agents.agent_toolkits import create_conversational_retrieval_agent

#Langchain given API to interact with chat models
from langchain_openai.chat_models import ChatOpenAI
llm = ChatOpenAI(openai_api_key=OPENAI_API_KEY,temperature=0)
agent_executor = create_conversational_retrieval_agent(llm,tools,verbose=True)
```

# **Invoking the LLM via the retriver tool**

```python
from IPython.display import Markdown, display
import textwrap
input = "What are the surprising power of atomic habbits"
#input  = "What are all the various chapters in this book"
result = agent_executor.invoke({"input": input})
#print(result["output"])


formatted_text = textwrap.fill(result['output'], width=100)
print(formatted_text)


#display(Markdown(result['output']))
```
